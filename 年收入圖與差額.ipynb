```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv('./範例檔/salary.csv', encoding='UTF-8')
```

```python
x = dat['年齡']
```

```python
y = dat['年收入']
```

```python

# 繪圖
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
# 資料筆數
cnt = len(dat)

# 取差額
diff_y = []
for i in range(0, cnt-1):
    diff_y.append(y[i+1] - y[i])

# 繪圖
plt.plot(x[1:], diff_y)
plt.grid(color='0.8')
plt.show()
```
